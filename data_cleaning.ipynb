```python
import numpy as np
import sampling_with_data_cleaning as sdc
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
vec1 = np.random.randn(50,2)
vec2 = 1+0.5*np.random.randn(10,2)
vec = np.concatenate((vec1,vec2),0)
print vec.shape
labels = np.concatenate((np.zeros((50,1)),np.ones((10,1))),0)
plt.scatter(vec[:,0],vec[:,1],c = labels)
```

```python
#import sampling_with_data_cleaning as sdc
labels = labels.reshape(labels.shape[0],)
x_new,y_new = sdc.SMOTE(vec,labels,h=5,alpha = 0.8,k=7)
plt.scatter(x_new[:,0],x_new[:,1],c=y_new)
```
